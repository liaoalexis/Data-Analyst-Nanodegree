## Introduction

WeRateDogs is a popular Twitter account that rates people's dogs with humorous comments. In this report, we present an analysis of the WeRateDogs dataset, which consists of over 2,000 tweets. After an extensive data wrangling process, we have a clean dataset to work with, and now we can uncover some interesting insights about the dogs rated on this platform.

### 1. Most Common Dog Breeds

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

df_cleaned = pd.read_csv('twitter_archive_master.csv')
breed_data = df_cleaned[['breed', 'retweets', 'likes']]
breed_grouped = breed_data.groupby(['breed']).mean()

# Sort the top 10 breeds by descending order of average likes
breed_grouped = breed_grouped.sort_values('likes', ascending=False).head(10)

fig, ax = plt.subplots(figsize=(10, 10))
breed_grouped.plot(kind='barh', ax=ax)
ax.set_xlabel('Count')
ax.set_ylabel('Breed')
ax.set_title('Average Retweets and Likes by Breed')
plt.show()
```

The top 10 most common dog breeds rated by WeRateDogs are:

Golden Retriever, Labrador Retriever, Pembroke Welsh Corgi, Chihuahua, Pug, Chow, Samoyed, Pomeranian, Toy Poodle, Alaskan Malamute

Golden Retrievers and Labrador Retrievers are the most frequently featured breeds, which is not surprising considering their popularity and lovable nature.

### 2. Popular Dog Stages

```python
stage_counts = df_cleaned['dog_stage'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(stage_counts, labels=stage_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Dog Stage Distribution')
plt.show()
```

WeRateDogs categorizes dogs into four stages: doggo, floofer, pupper, and puppo. Among these, the pupper stage is the most popular, accounting for a significant portion of the dataset. This popularity could be due to the cute and playful nature of younger dogs.

### 3. Dog Breeds with the Highest Retweet and Favorite Counts on Average

```python
breed_data = df_cleaned[['breed', 'retweets', 'likes']]
breed_grouped = breed_data.groupby(['breed']).mean()

# Sort the top 10 breeds by descending order of average likes
breed_grouped = breed_grouped.sort_values('likes', ascending=False).head(10)

fig, ax = plt.subplots(figsize=(10, 10))
breed_grouped.plot(kind='barh', ax=ax)
ax.set_xlabel('Count')
ax.set_ylabel('Breed')
ax.set_title('Average Retweets and Likes by Breed')
plt.show()
```

We also analyzed the average retweet and favorite counts for each breed. Salukis have the highest average likes, while Standard Poodles have the highest average retweets.
